```python
%config Completer.use_jedi = False
%matplotlib notebook

import xarray as xr
import netCDF4 as nc4

nc = nc4.Dataset('u_wind.nc', mode='r')  # Or from siphon.ncss
dataset = xr.open_dataset(xr.backends.NetCDF4DataStore(nc))
```

```python
dataset
```

```python
data_timeavg=dataset.mean(dim='time')
```

```python
data_timelonavg=data_timeavg.mean(dim='longitude')
```

```python
import matplotlib.pyplot as plt

plt.contourf(data_timelonavg['latitude'],data_timelonavg['level'],data_timelonavg['u'])
plt.gca().invert_yaxis()
plt.colorbar()
```
